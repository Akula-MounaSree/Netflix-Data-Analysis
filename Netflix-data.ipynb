```python
#numpy operations
#Step 1: Extract only numbers from duration
import numpy as np
df['duration_clean']=(df['duration'].str.extract('(\d+)').astype(float))
#Step 2: Store those numeric values in an array
arr=df['duration_clean'].dropna().to_numpy()
#mean, median, quantile
mean_val=np.mean(arr)
median_val=np.median(arr)
q1,q3=np.percentile(arr,[25,75])
# quartile is used to divide into 4 segment to understand the data's distribution, centre, and data spread and outliers means identifying the low or high values
#IQR = q3 - q1 is interquartile shows the range of middle 50% of data
iqr=q3-q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

df['duration_capped']=np.where(df['duration_clean']>upper,upper,
                              np.where(df['duration_clean']<lower, lower, df['duration_clean']))
df[['duration', 'duration_clean', 'duration_capped']].head(10)
```
